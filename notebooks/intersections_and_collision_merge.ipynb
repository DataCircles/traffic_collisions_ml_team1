```python
#Load the data
collisions_df = pd.read_pickle('../data/processed/cleaned_data.pkl')
intersections_df = pd.read_csv('../data/Intersections.csv')
```

```python
int_collisions = collisions_df[collisions_df.ADDRTYPE == 'Intersection'].copy()
```

```python
int_collisions.rename(columns={'INTKEY':'COMPKEY'}, inplace=True)
```
